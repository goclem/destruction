## Preparing shapefiles 

The following functions are used to transform raw damage, settlement, and no-analysis shapefiles into the format needed by our architecture.

````python
import sys
import warnings
warnings.filterwarnings('ignore')

# setting path
sys.path.append('../')
  
from destruction_utilities import prep_all, search_data

cities = ['aleppo', 'damascus', 'daraa', 'deir-ez-zor','hama', 'homs', 'idlib', 'raqqa']
for city in cities:
    prep_all(city, '', '../../data')
````

## Documentation
---
### prep_damage(city:str, suffix:str, datadir:str)
Function to take raw damage shapefile and preprocess it to get to the format needed by our architecture (see below).

Damage

    Name: [city]_damage.gpkg
    Variables (integer):
    Names: [yyyy-mm-dd] (one per census)
    Values:
        1 low destruction
        2 medium destruction
        3 high destruction
        0 otherwise

#### Usage
```
prep_damage("aleppo", "test", "../../data")
```
#### Output

The above function call Will create file "../../data/aleppo/others/allepo_damage_test.gpkg" with the above specs.

---
### prep_settlement(city:str, suffix:str, datadir:str)
Function to take raw settlement shapefile and preprocess it to get to the format needed by our architecture (see below).

Settlement

    Name: [city]_settlement.gpkg
    Variables: None

#### Usage
```
prep_settlement("aleppo", "test", "../../data")
```
#### Output

The above function call Will create file "../../data/aleppo/others/allepo_settlement_test.gpkg" with the above specs.

---
### prep_noanalysis(city:str, suffix:str, datadir:str)
Function to take raw no-analysis shapefile and preprocess it to get to the format needed by our architecture (see below).

Noanalysis

    Name: [city]_noanalysis.gpkg
    Variables (string): reason

#### Usage
```
prep_noanalysis("aleppo", "test", "../../data")
```
#### Output

The above function call Will create file "../../data/aleppo/others/allepo_noanalysis_test.gpkg" with the above specs.


---
### prep_all(city:str, suffix:str, datadir:str)
Wrapper function that calls the above three functions all at once for a given city.
