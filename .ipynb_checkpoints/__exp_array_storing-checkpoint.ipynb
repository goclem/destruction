```python
from destruction_utilities import *
import zarr
```

```python
CITY = 'aleppo'

img_path = f'../data/{CITY}/others/{CITY}_images.zarr'
label_path = f'../data/{CITY}/others/{CITY}_labels.zarr'

images  = search_data(pattern(city=CITY, type='image'))
labels  = search_data(pattern(city=CITY, type='label'))

samples_path = f'../data/{CITY}/others/{CITY}_samples.tif'
samples = read_raster(samples_path)
```

```python
temp = np.array(read_raster(images[0]))
w, h, b = temp.shape
temp = temp.reshape(1,w,h, b)
temp = tile_sequences(temp,  tile_size=(128, 128))
```

```python
noanalysis, train, validate, test = sample_split(temp, samples.flatten())
```

```python
for image in images:
    temp = np.array(read_raster(image))
    w, h, b = temp.shape
    temp = temp.reshape(1,w,h, b)
    temp = tile_sequences(temp,  tile_size=(128, 128))
    p, i, w, h, b = temp.shape
    temp = temp.reshape(p*i, h, w, b)
    if not exists(img_path):
        print('Saving image 0 to disk')
        zarr.save(img_path, temp)        
    else:
        print('Appending to file on disk')
        za = zarr.open(img_path, mode='a')
        za.append(temp)
    del temp, p, i, w, h, b
```

```python
for label in labels:
    temp = np.array(read_raster(label))
    w, h, b = temp.shape
    temp = temp.reshape(1,w,h, b)
    temp = tile_sequences(temp,  tile_size=(1, 1))
    p, i, w, h, b = temp.shape
    temp = temp.reshape(p*i, h, w, b)
    if not exists(label_path):
        print('Saving label 0 to disk')
        zarr.save(label_path, temp)        
    else:
        print('Appending to file on disk')
        za = zarr.open(label_path, mode='a')
        za.append(temp)
    del temp
```

```python
class ZarrGenerator(keras.utils.all_utils.Sequence):
    def __init__(self, city, batch_size, dim, n_channels, n_labels, shuffle):
        self.city = city
        self.batch_size = batch_size
        self.n_channels = n_channels
        self.n_labels = n_labels
        self.dim = dim
        self.shuffle = shuffle
        self._connect()
        
    def _connect(self):
        img_path = f'../data/{self.city}/others/{self.city}_images.zarr'
        label_path = f'../data/{self.city}/others/{self.city}_labels.zarr'
        self.images = zarr.open(img_path, mode='r')
        self.labels = zarr.open(label_path, mode='r')
        self.indexes = np.arange(len(self.images))
        
    def __len__(self):
        return len(self.images)//self.batch_size
    
    def __getitem__(self, index):
        indexes = self.indexes[index*self.batch_size:(index+1)*self.batch_size]
        
        X = np.empty((self.batch_size, *self.dim, self.n_channels))
        y = np.empty((self.batch_size), dtype=int)
                
        for i, idx in enumerate(indexes):
            X[i,] = self.images[idx]
            y[i,] = self.labels[idx]
        
        return X,y
    
    def __data_generation(self, indexes):
        

        print("ooh")
        
    def on_epoch_end(self):
        'Updates indexes after each epoch'
        self.indexes = np.arange(len(self.images))
        if self.shuffle == True:
            np.random.shuffle(self.indexes)
```
